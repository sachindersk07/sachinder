```python
import pandas as pd
import numpy as np
df=pd.read_excel("C:\\Users\\ignite438\\Documents\\college_data.xlsx")
df
```

```python

co=df.loc[df['Gender Mix'].isin(['Co-Ed'])]
wo=df.loc[df['Gender Mix'].isin(['Womens'])]
me=df.loc[df['Gender Mix'].isin(['Mens'])]

cocount= sum(df['Gender Mix'] == 'Co-Ed')
wocount=sum(df['Gender Mix'] == 'Womens')
mencount=sum(df['Gender Mix'] == 'Mens')

import matplotlib.pyplot as plt

labels = ["Co-Ed","Women","Mens"]

explode_array = [0,0,0]

colours_array = ["#ffff99","#fdc086","#beaed4"]

counts=[cocount,wocount,mencount]

plt.pie(counts,colors = colours_array,autopct='%.2f%%',startangle=90)

plt.axis('equal')

plt.title('the OI college distribution by Gender Mix', fontsize=16)

plt.legend(labels,bbox_to_anchor=(0.9,0.6))
plt.show()
```

```python
gov=df.loc[df['Institution Type'].isin(['Government'])]
pr=df.loc[df['Institution Type'].isin(['Private'])]

govcount=sum(df['Institution Type'] == 'Government')
prcount=sum(df['Institution Type'] == 'Private')
```
